```python
import pandas as pd, ipywidgets as widgets, numpy as np, matplotlib.pyplot as plt, io
uploader = widgets.FileUpload(multiple = True)
display(uploader)
```

```python
files_dict = uploader.value
files_dict
```

```python
data_list = []
for excel_sheet in files_dict.keys():
    this_data = pd.read_excel(io = io.BytesIO(files_dict[excel_sheet]['content']), index_col=0)
    data_list.append(this_data)
all_data = pd.concat(data_list, ignore_index = True)
all_data.index += 1
```

```python
data_cleaned = all_data.drop(columns = ['iQue3 Column', 'Well Type', 'Transfection cell line', 'Purification Method', 'Transfection scale (mL)', 'Volume (uL)', 'Format Row', 'Format Column', 'Count of singlets', 'Date Delivered'], inplace = False)
data_cleaned.to_excel("All Data Compiled Cleaned.xlsx", sheet_name = "All Data")
data_cleaned
```

```python
BPP_614 = data_cleaned[data_cleaned["BPP Ticket# (If available)"] == "BPP-614"]
BPP_616 = data_cleaned[data_cleaned["BPP Ticket# (If available)"] == "BPP-616"]
controls = data_cleaned[data_cleaned["BPP Ticket# (If available)"] == "Control"]
BPP_614 = pd.concat([BPP_614, controls])
BPP_616 = pd.concat([BPP_616, controls])
BPP_614.reset_index(drop = True, inplace = True)
BPP_616.reset_index(drop = True, inplace = True)
controls.reset_index(drop = True, inplace = True)
BPP_614.index += 1
BPP_616.index += 1
controls.index += 1
BPP_614_cleaned = BPP_614[["Sample Name (TB/SC,etc.)", "A2a MFI Ratio", "A2b MFI Ratio"]]
BPP_614_cleaned.rename(columns = {'Sample Name (TB/SC,etc.)':'Antibody Name'}, inplace = True)
BPP_614_cleaned.to_excel("BPP-614_cleaned.xlsx")
BPP_616_cleaned = BPP_616[["Sample Name (TB/SC,etc.)", "A2a MFI Ratio", "A2b MFI Ratio"]]
BPP_616_cleaned.rename(columns = {'Sample Name (TB/SC,etc.)':'Antibody Name'}, inplace = True)
BPP_616_cleaned.to_excel("BPP-616_cleaned.xlsx")
#BPP_614.sort_values(by = "A2a MFI Ratio", ascending = False)
```

```python
BPP_614.to_excel("BPP-614.xlsx", sheet_name = "BPP-614")
BPP_616.to_excel("BPP-616.xlsx", sheet_name = "BPP-616")
```

```python
def my_plotter(ax, data, controls, bpp):
    ax.scatter(x = data.index, y = data["A2a MFI Ratio"], alpha = .2, color = "blue")
    ax.set_title(bpp +" Primary Screen")
    ax.set_xlabel("Antibody Number")
    ax.set_ylabel("A2a MFI Ratio")
    ax.scatter(x = controls.index, y = controls["A2a MFI Ratio"], alpha = .5, color = "red", label = "Controls")
    ax.legend()
    
fig, ax = plt.subplots()
my_plotter(ax, BPP_614, controls, "BPP-614")
```
