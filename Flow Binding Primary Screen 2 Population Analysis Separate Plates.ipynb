```python
iquename = list(raw_uploader.value.keys())[0]
platemap1_name, platemap2_name = list(platemap1_uploader.value.keys())[0], list(platemap2_uploader.value.keys())[0]
iquedata = pd.read_csv(io.BytesIO(raw_uploader.value[iquename]['content']))
platemap1 = pd.read_excel(io.BytesIO(platemap1_uploader.value[platemap1_name]['content']), sheet_name = "Protein Quant")
platemap2 = pd.read_excel(io.BytesIO(platemap2_uploader.value[platemap2_name]['content']), sheet_name = "Protein Quant")
first_population_name = first_population_widget.value
second_population_name = second_population_widget.value
iquedata.rename(columns = {'Plate' : 'iQue3 Plate','Well ID':'iQue3 384 Well ID', 'Column':'iQue3 Column'}, 
                inplace = True)
platemap1.rename(columns = {'Plate' : 'Antibody Plate','Well ID':'Antibody Platemap 96 Well ID'}, inplace = True)
platemap2.rename(columns = {'Plate' : 'Antibody Plate','Well ID':'Antibody Platemap 96 Well ID'}, inplace = True)
iquedata
```

```python
data_cleaned = iquedata[["Experiment", "Export Date and Time", "iQue3 Plate", "iQue3 384 Well ID", "iQue3 Column", "Well Type", "Median RL1-H of singlets", "Count of singlets"]]
data_cleaned
```

```python
#Return data frame with only samples of the well type specified
def clean_for_well_type(well_type_string):
    return data_cleaned[data_cleaned["Well Type"] == well_type_string]

#Return data frame with only odd or even columns, depending on the remainder given, and of a given plate type
#"remainder": when column is modulo by 2, so 0 are even columns, 1 are odd columns
#plate_type: the letter denoting what population is in the plate, should be the first letter of the plate string
def clean_for_well_type(remainder, plate_type):
    ret_data = data_cleaned[data_cleaned["iQue3 Column"] % 2 == remainder]
    ret_data = ret_data[ret_data["iQue3 Plate"].str.slice(start = 0, stop = 1) == plate_type]
    return ret_data

#Average function that takes a string indicating the well type of interest (marked in iQue3 software, or manually changed
#Separates n=1 and n=2 replicates from 384 well format, resets indeces, and averages
#Returns a series with the averaged MFI ratios, removing the replicate wells
def average_MFI(well_type_string):
    selected_wells = clean_for_well_type(well_type_string)
    MFIs = selected_wells["Median RL1-H of singlets"]
    firsts = MFIs.iloc[::2]
    seconds = MFIs.iloc[1::2]
    firsts.reset_index(inplace = True, drop = True)
    seconds.reset_index(inplace = True, drop = True)
    average = firsts + seconds
    average = average/2
    return average

#Average function that takes a remainder and a plate_type to pass to clean_for_well_type
#Separates n=1 and n=2 replicates from 384 well format, resets indeces, and averages
#Returns a series with the averaged MFI ratios, removing the replicate wells
def average_MFI(remainder, plate_type):
    selected_wells = clean_for_well_type(remainder, plate_type)
    MFIs = selected_wells["Median RL1-H of singlets"]
    firsts = MFIs.iloc[::2]
    seconds = MFIs.iloc[1::2]
    firsts.reset_index(inplace = True, drop = True)
    seconds.reset_index(inplace = True, drop = True)
    average = firsts + seconds
    average = average/2
    return average

#Calculates the MFI ratio given a positive and negative series, 
#and appends the MFI ratio to the original data frame, excluding the negative population
#first_pop_series: the series of the first population values
#second_pop_series: the series of the second population values
#negative_series: the series of the untransfected population values
#first_population_string: the string that marks the positive samples in the Well Type column from iQue3 csv
#second_population_string: name of the second population
def calculate_analysis_sheet_2pop(first_pop_series, second_pop_series, negative_series, first_population_string, 
                                  second_population_string):
    first_pop_MFI_ratio = first_pop_series/negative_series
    second_pop_MFI_ratio = second_pop_series/negative_series
    original_positive = clean_for_well_type(first_population_string)
    original_positive = original_positive[::2]
    original_positive.reset_index(inplace = True, drop = True)
    original_positive[first_population_string + " MFI Ratio"] = first_pop_MFI_ratio
    original_positive[first_population_string + " Averaged Raw MFI"] = first_pop_series
    original_positive[second_population_string + " MFI Ratio"] = second_pop_MFI_ratio
    original_positive[second_population_string + " Averaged Raw MFI"] = second_pop_series
    original_positive["Parental Averaged Raw MFI"] = negative_series
    analysis_sheet = pd.concat([original_positive, ab_sheet], axis = 1)
    analysis_sheet.index += 1
    analysis_sheet.drop(columns = ['Median RL1-H of singlets'], inplace = True)
    return analysis_sheet

#Calculates the MFI ratio given a positive and negative series, 
#and appends the MFI ratio to the original data frame, excluding the negative population
#first_pop_series: the series of the first population values
#second_pop_series: the series of the second population values
#negative_series: the series of the untransfected population values
#first_population_string: the string that marks the positive samples in the Well Type column from iQue3 csv
#second_population_string: name of the second population
#plate_num: 1 indicates the odd columns, platemap1. 2 indicates even columns, platemap2.
def calculate_analysis_sheet_2pop(first_pop_series, second_pop_series, negative_series, first_population_string, 
                                  second_population_string, plate_num, plate_letter, platemap):
    first_pop_MFI_ratio = first_pop_series/negative_series
    second_pop_MFI_ratio = second_pop_series/negative_series
    original_positive = clean_for_well_type(plate_num%2, plate_letter)
    original_positive = original_positive[::2]
    original_positive.reset_index(inplace = True, drop = True)
    original_positive[first_population_string + " Averaged Raw MFI"] = first_pop_series
    original_positive[second_population_string + " Averaged Raw MFI"] = second_pop_series
    original_positive["Parental Averaged Raw MFI"] = negative_series
    original_positive[first_population_string + " MFI Ratio"] = first_pop_MFI_ratio
    original_positive[second_population_string + " MFI Ratio"] = second_pop_MFI_ratio
    analysis_sheet = pd.concat([original_positive, platemap], axis = 1)
    analysis_sheet.index += 1
    analysis_sheet.drop(columns = ['Median RL1-H of singlets', 'Unnamed: 0'], inplace = True)
    return analysis_sheet
```
